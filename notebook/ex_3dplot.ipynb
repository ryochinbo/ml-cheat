```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# ランダムデータの生成
np.random.seed(0)  # 再現性のためにシードを設定
data_size = 15
x = np.random.randint(0, 51, data_size)
y = np.random.randint(0, 51, data_size)
z = np.random.randint(0, 51, data_size)
a = np.random.randint(0, 51, data_size)
labels = np.random.choice(['try1', 'try2', 'try3'], data_size)
df = pd.DataFrame({'x': x, 'y': y, 'z': z, 'a': a, 'label': labels})

```

```python
df.head()
```

```python
# 3次元散布図の作成
fig = plt.figure(figsize=(12, 12))  # 図のサイズを指定
ax = fig.add_subplot(111, projection='3d')

# ラベル列を追加
labels = np.random.choice(['try1', 'try2', 'try3'], data_size)
df = pd.DataFrame({'x': x, 'y': y, 'z': z, 'label': labels})

# 散布図の作成
sc = ax.scatter(df['x'], df['y'], df['z'], s=50)

# 軸ラベルの設定
ax.set_xlabel('Xline(mm)')
ax.set_ylabel('yline(g)')
ax.set_zlabel('zline(ml)')

# 軸の上下限を設定
ax.set_xlim(0, 50)  # x軸の範囲
ax.set_ylim(0, 50)  # y軸の範囲
ax.set_zlim(0, 50)  # z軸の範囲
ax.invert_yaxis() #<----軸を反転する

# プロットの表示
plt.show()
```

```python
# 3次元散布図の作成
# 3次元散布図の作成
fig = plt.figure(figsize=(12, 12))  # 図のサイズを指定
ax = fig.add_subplot(111, projection='3d')

# ラベルごとに色を設定
colors = {'try1': 'r', 'try2': 'g', 'try3': 'b'}  # 色の辞書
for label in colors.keys():
    subset = df[df['label'] == label]
    ax.scatter(subset['x'], subset['y'], subset['z'], color=colors[label], label=label, s=70)


# # 散布図の作成
# sc = ax.scatter(df['x'], df['y'], df['z'], s=50)

# 軸ラベルの設定
ax.set_xlabel('Xline(mm)')
ax.set_ylabel('yline(g)')
ax.set_zlabel('zline(ml)')

# 軸の上下限を設定
ax.set_xlim(0, 50)  # x軸の範囲
ax.set_ylim(0, 50)  # y軸の範囲
ax.set_zlim(0, 50)  # z軸の範囲
ax.invert_yaxis() #<----軸を反転する


# 凡例の表示
ax.legend(title='Label')

# プロットの表示
plt.show()
```

```python
# 3次元散布図の作成
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 散布図の作成
sc = ax.scatter(df['x'], df['y'], df['z'], c=df['a'], cmap='viridis')
# カラーバーの追加
cbar = plt.colorbar(sc, ax=ax, pad=0.1)  # padでカラーバーの
cbar.set_label('a_score')
# 軸ラベルの設定
ax.set_xlabel('Xline(mm)')
ax.set_ylabel('yline(g)')
ax.set_zlabel('zline(ml)')
# プロットの表示
plt.show()
```
